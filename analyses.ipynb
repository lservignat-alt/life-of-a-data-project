```python
proj_daymet = "+proj=lcc +lat_1=46.8 +lat_0=46.8 +lon_0=0 +k_0=0.99987742 +x_0=600000 +y_0=2200000 +a=6378249.2 +b=6356515 +towgs84=-168,-60,320,0,0,0,0 +pm=paris +units=m +no_defs"
P = pyproj.Proj(proj_daymet)
G = pyproj.Geod(proj_daymet)

def LatLon_To_XY(Lat,Lon):
    return P(Lat,Lon)

def XY_To_LatLon(x,y):
    return P(x,y,inverse=True)    

def distance(Lat1, Lon1, Lat2, Lon2):
    return G.inv(Lon1, Lat1, Lon2, Lat2)[2]
```

```python
# =============================================================================
# NETTOYAGE DES DONNEES - TELECOM - GROUPE 5 SUD-EST
# =============================================================================
# Basé sur le dictionnaire réel : Dico_Donnees.xlsx
#
# AVANT DE LANCER CE SCRIPT :
#   1. Installe les librairies si nécessaire :
#      pip install pandas openpyxl pyproj
#   2. Mets les 3 fichiers CSV dans le même dossier que ce script
#   3. Vérifie les paramètres dans la section CONFIGURATION ci-dessous
# =============================================================================

import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")


# =============================================================================
# CONFIGURATION — A ADAPTER SI BESOIN
# =============================================================================

# !! MODIFIE CES CHEMINS selon où sont tes fichiers !!
FICHIER_PRINCIPAL = "Fichier_principal_pedago_anonym.csv"
FICHIER_COORD     = "COORD_BQT.csv"
FICHIER_LIB       = "LIB_BQT.csv"

# Séparateur CSV — change en ";" si tes fichiers utilisent le point-virgule
SEPARATEUR = ","

# Encodage — essaie "latin-1" si tu as des erreurs de caractères spéciaux
ENCODAGE = "utf-8"

# Filtre groupe 5
ZONE_CIBLE = "5_SUD_EST"

# Années de naissance valides pour un client
AGE_MIN = 1930   # clients nés après 1930
AGE_MAX = 2006   # clients nés avant 2006 (18 ans minimum)

# Types de produits valides (à compléter si tu vois d'autres valeurs dans tes données)
LINE_TYPE_VALIDES = ["T", "M", "I"]  # T=Terminal, M=Mobile, I=Internet

# Emplacements valides dans COORD_BQT
EMPLACEMENT_VALIDES = ["CV", "CC"]   # CV=Centre-Ville, CC=Centre-Commercial


# =============================================================================
# UTILITAIRES
# =============================================================================

journal = []  # on garde une trace de toutes les décisions

def log(message):
    """Affiche et enregistre un message dans le journal."""
    print(f"  >> {message}")
    journal.append(message)

def separateur(titre):
    print(f"\n{'='*65}")
    print(f"  {titre}")
    print(f"{'='*65}")

def nb_lignes(df, label=""):
    msg = f"{label} : {len(df):,} lignes"
    log(msg)
    return len(df)


# =============================================================================
# ETAPE 1 — CHARGEMENT DES FICHIERS
# =============================================================================
separateur("ETAPE 1 — CHARGEMENT DES FICHIERS")

# --- Fichier principal ---
print("\n[1.1] Chargement du fichier principal...")
try:
    df = pd.read_csv(FICHIER_PRINCIPAL, sep=SEPARATEUR, encoding=ENCODAGE, low_memory=False)
    log(f"Fichier principal charge : {len(df):,} lignes x {len(df.columns)} colonnes")
    log(f"Colonnes : {list(df.columns)}")
except Exception as e:
    print(f"  ERREUR chargement fichier principal : {e}")
    print("  => Essaie de changer SEPARATEUR en ';' ou ENCODAGE en 'latin-1'")
    raise

# --- COORD_BQT ---
print("\n[1.2] Chargement de COORD_BQT...")
try:
    coord = pd.read_csv(FICHIER_COORD, sep=SEPARATEUR, encoding=ENCODAGE)
    log(f"COORD_BQT charge : {len(coord):,} lignes x {len(coord.columns)} colonnes")
    log(f"Colonnes : {list(coord.columns)}")
except Exception as e:
    print(f"  ERREUR : {e}")
    raise

# --- LIB_BQT ---
print("\n[1.3] Chargement de LIB_BQT...")
try:
    lib = pd.read_csv(FICHIER_LIB, sep=SEPARATEUR, encoding=ENCODAGE)
    log(f"LIB_BQT charge : {len(lib):,} lignes x {len(lib.columns)} colonnes")
    log(f"Colonnes : {list(lib.columns)}")
except Exception as e:
    print(f"  ERREUR : {e}")
    raise


# =============================================================================
# ETAPE 2 — APERCU RAPIDE (diagnostic initial)
# =============================================================================
separateur("ETAPE 2 — DIAGNOSTIC INITIAL")

print("\n[2.1] Types de colonnes — Fichier principal :")
print(df.dtypes.to_string())

print("\n[2.2] Valeurs manquantes — Fichier principal :")
nulls = df.isnull().sum()
nulls_pct = (nulls / len(df) * 100).round(2)
diag = pd.DataFrame({"Nulls": nulls, "Pct (%)": nulls_pct})
diag = diag[diag["Nulls"] > 0].sort_values("Pct (%)", ascending=False)
if diag.empty:
    print("  Aucune valeur manquante !")
else:
    print(diag.to_string())

print("\n[2.3] Valeurs uniques par colonne catégorielle (fichier principal) :")
for col in ["LINE_TYPE", "ZONE", "Geolife", "Geolife_AGG", "NET_ZONE_IN",
            "DENSE_ZONE_IN", "PERSON_SALUTATION_CD"]:
    if col in df.columns:
        vals = df[col].value_counts(dropna=False).head(10)
        print(f"\n  {col} :")
        print(vals.to_string())

print("\n[2.4] Valeurs uniques — COORD_BQT :")
if "Emplacement_PDV" in coord.columns:
    print("  Emplacement_PDV :", coord["Emplacement_PDV"].value_counts(dropna=False).to_dict())
if "Surface_COMMERC" in coord.columns:
    print(f"  Surface_COMMERC : min={coord['Surface_COMMERC'].min()}, max={coord['Surface_COMMERC'].max()}")

print("\n[2.5] Valeurs uniques — LIB_BQT :")
if "MKT_CHANNL_LB" in lib.columns:
    print("  MKT_CHANNL_LB :", lib["MKT_CHANNL_LB"].value_counts(dropna=False).to_dict())

print("\n[2.6] Distribution PERIOD_YYYY et PERIOD_MM :")
if "PERIOD_YYYY" in df.columns:
    print("  Années :", sorted(df["PERIOD_YYYY"].dropna().unique().tolist()))
if "PERIOD_MM" in df.columns:
    print("  Mois   :", sorted(df["PERIOD_MM"].dropna().unique().tolist()))


# =============================================================================
# ETAPE 3 — NETTOYAGE DU FICHIER PRINCIPAL
# =============================================================================
separateur("ETAPE 3 — NETTOYAGE DU FICHIER PRINCIPAL")
n_initial = len(df)
log(f"Nb lignes initial : {n_initial:,}")

# ── 3.1 Suppression des doublons ─────────────────────────────────────────────
print("\n[3.1] Suppression des doublons...")
n_avant = len(df)
df = df.drop_duplicates()
n_suppr = n_avant - len(df)
log(f"Doublons supprimes : {n_suppr:,} lignes")

# ── 3.2 Filtre ZONE = 5_SUD_EST ─────────────────────────────────────────────
print(f"\n[3.2] Filtre sur ZONE = '{ZONE_CIBLE}'...")
n_avant = len(df)
df = df[df["ZONE"] == ZONE_CIBLE].copy()
n_suppr = n_avant - len(df)
log(f"Lignes hors SUD-EST supprimees : {n_suppr:,} | Reste : {len(df):,}")

if len(df) == 0:
    print("  ATTENTION : 0 lignes apres filtre ZONE !")
    print(f"  Valeurs disponibles dans ZONE : {df['ZONE'].unique()}")
    print("  => Verifie la valeur exacte dans ZONE_CIBLE")
    raise ValueError("Filtre ZONE vide — ajuste ZONE_CIBLE")

# ── 3.3 Nettoyage LINE_TYPE ──────────────────────────────────────────────────
print("\n[3.3] Nettoyage LINE_TYPE...")
df["LINE_TYPE"] = df["LINE_TYPE"].str.strip().str.upper()
n_avant = len(df)
df = df[df["LINE_TYPE"].isin(LINE_TYPE_VALIDES)]
n_suppr = n_avant - len(df)
log(f"LINE_TYPE invalides supprimes : {n_suppr:,} lignes | Valides : {LINE_TYPE_VALIDES}")

# ── 3.4 Nettoyage PERIOD_YYYY et PERIOD_MM ──────────────────────────────────
print("\n[3.4] Nettoyage et validation des dates (PERIOD_YYYY / PERIOD_MM)...")

# Convertir en entiers si nécessaire
df["PERIOD_YYYY"] = pd.to_numeric(df["PERIOD_YYYY"], errors="coerce")
df["PERIOD_MM"]   = pd.to_numeric(df["PERIOD_MM"],   errors="coerce")

# Vérifier les années raisonnables (entre 2010 et 2026)
n_avant = len(df)
df = df[df["PERIOD_YYYY"].between(2010, 2026)]
log(f"Annees hors [2010-2026] supprimees : {n_avant - len(df):,} lignes")

# Vérifier les mois valides (1 à 12)
n_avant = len(df)
df = df[df["PERIOD_MM"].between(1, 12)]
log(f"Mois invalides supprimes : {n_avant - len(df):,} lignes")

# Créer une colonne date combinée (utile pour les graphiques)
df["DATE"] = pd.to_datetime(
    df["PERIOD_YYYY"].astype(int).astype(str) + "-" +
    df["PERIOD_MM"].astype(int).astype(str).str.zfill(2) + "-01",
    errors="coerce"
)
log("Colonne DATE creee (format YYYY-MM-01)")

# ── 3.5 Nettoyage PERSON_BIRTH_DT_Year ──────────────────────────────────────
print("\n[3.5] Nettoyage PERSON_BIRTH_DT_Year (annee de naissance)...")
df["PERSON_BIRTH_DT_Year"] = pd.to_numeric(df["PERSON_BIRTH_DT_year"], errors="coerce")
n_avant = len(df)
df = df[df["PERSON_BIRTH_DT_Year"].between(AGE_MIN, AGE_MAX) | df["PERSON_BIRTH_DT_Year"].isnull()]
n_suppr = n_avant - len(df)
log(f"Annees de naissance aberrantes supprimees : {n_suppr:,} lignes | Plage : [{AGE_MIN}-{AGE_MAX}]")

# ── 3.6 Nettoyage ORDER_SHOP_CD (clé de jointure) ───────────────────────────
print("\n[3.6] Nettoyage ORDER_SHOP_CD (cle de jointure)...")
df["ORDER_SHOP_CD"] = df["ORDER_SHOP_CD"].astype(str).str.strip()
n_avant = len(df)
df = df[df["ORDER_SHOP_CD"].notna() & (df["ORDER_SHOP_CD"] != "") & (df["ORDER_SHOP_CD"] != "nan")]
n_suppr = n_avant - len(df)
log(f"ORDER_SHOP_CD vides/nuls supprimes : {n_suppr:,} lignes")

# ── 3.7 Nettoyage CITY_LN ────────────────────────────────────────────────────
print("\n[3.7] Nettoyage CITY_LN (ville du client)...")
if "CITY_LN" in df.columns:
    df["CITY_LN"] = df["CITY_LN"].astype(str).str.strip().str.upper()
    df["CITY_LN"] = df["CITY_LN"].replace({"NAN": np.nan, "": np.nan})
    n_nulls = df["CITY_LN"].isnull().sum()
    log(f"CITY_LN normalise en majuscules | Valeurs nulles restantes : {n_nulls:,}")

# ── 3.8 Nettoyage Geolife ────────────────────────────────────────────────────
print("\n[3.8] Nettoyage Geolife...")
for col in ["Geolife", "Geolife_AGG"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace({"nan": np.nan, "": np.nan})
        n_nulls = df[col].isnull().sum()
        log(f"{col} : {df[col].nunique()} categories uniques | {n_nulls:,} valeurs nulles")

# ── 3.9 Nettoyage NET_ZONE_IN et DENSE_ZONE_IN ──────────────────────────────
print("\n[3.9] Nettoyage NET_ZONE_IN et DENSE_ZONE_IN...")
for col in ["NET_ZONE_IN", "DENSE_ZONE_IN"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace({"nan": np.nan, "": np.nan})
        log(f"{col} : {df[col].value_counts(dropna=False).to_dict()}")

# ── 3.10 Nettoyage DEPRTMNT_ID et POSTAL_CD ──────────────────────────────────
print("\n[3.10] Nettoyage DEPRTMNT_ID et POSTAL_CD...")
for col in ["DEPRTMNT_ID", "POSTAL_CD"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.zfill(2 if col == "DEPRTMNT_ID" else 5)
        df[col] = df[col].replace({"nan": np.nan, "00": np.nan, "00000": np.nan})
        log(f"{col} nettoye | {df[col].isnull().sum():,} valeurs nulles")

# ── 3.11 Nettoyage coordonnées client (Lambert II étendu) ────────────────────
print("\n[3.11] Nettoyage Coord_L2E_II_X / COORD_Y (coordonnees client Lambert II)...")
for col in ["Coord_L2E_II_X", "Coord_L2E_II_Y"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
n_nulls_x = df["Coord_L2E_II_X"].isnull().sum() if "Coord_L2E_II_Y" in df.columns else 0
n_nulls_y = df["Coord_L2E_II_Y"].isnull().sum() if "Coord_L2E_II_Y" in df.columns else 0
log(f"Coord_L2E_II_X nulls : {n_nulls_x:,} | Coord_L2E_II_Y nulls : {n_nulls_y:,}")

log(f"Nb lignes apres nettoyage fichier principal : {len(df):,}")


# =============================================================================
# ETAPE 4 — NETTOYAGE COORD_BQT
# =============================================================================
separateur("ETAPE 4 — NETTOYAGE COORD_BQT")

# ── 4.1 Doublons ─────────────────────────────────────────────────────────────
n_avant = len(coord)
coord = coord.drop_duplicates(subset=["ORDER_SHOP_CD"])
log(f"Doublons ORDER_SHOP_CD supprimes dans COORD_BQT : {n_avant - len(coord):,}")

# ── 4.2 Nettoyage clé de jointure ────────────────────────────────────────────
coord["ORDER_SHOP_CD"] = coord["ORDER_SHOP_CD"].astype(str).str.strip()

# ── 4.3 Nettoyage Emplacement_PDV ────────────────────────────────────────────
coord["Emplacement_PDV"] = coord["Emplacement_PDV"].astype(str).str.strip().str.upper()
coord["Emplacement_PDV"] = coord["Emplacement_PDV"].replace({"NAN": np.nan, "": np.nan})
n_invalides = (~coord["Emplacement_PDV"].isin(EMPLACEMENT_VALIDES) & coord["Emplacement_PDV"].notna()).sum()
log(f"Emplacement_PDV : {coord['Emplacement_PDV'].value_counts(dropna=False).to_dict()}")
log(f"Valeurs hors [CC, CV] : {n_invalides}")

# ── 4.4 Nettoyage Surface_COMMERC ────────────────────────────────────────────
coord["Surface_COMMERC"] = pd.to_numeric(coord["Surface_COMMERC"], errors="coerce")
n_nulls = coord["Surface_COMMERC"].isnull().sum()
n_negatifs = (coord["Surface_COMMERC"] <= 0).sum()
coord.loc[coord["Surface_COMMERC"] <= 0, "Surface_COMMERC"] = np.nan
log(f"Surface_COMMERC : {n_nulls} nulls | {n_negatifs} valeurs <= 0 (mises a NaN)")

# ── 4.5 Nettoyage coordonnées boutiques (Lambert II étendu) ──────────────────
for col in ["Coord_L2E_II_X", "Coord_L2E_II_Y"]:
    coord[col] = pd.to_numeric(coord[col], errors="coerce")
n_sans_coord = coord[["Coord_L2E_II_X", "Coord_L2E_II_Y"]].isnull().any(axis=1).sum()
log(f"Boutiques sans coordonnees : {n_sans_coord}")

log(f"COORD_BQT apres nettoyage : {len(coord):,} boutiques")


# =============================================================================
# ETAPE 5 — NETTOYAGE LIB_BQT
# =============================================================================
separateur("ETAPE 5 — NETTOYAGE LIB_BQT")

# ── 5.1 Doublons ─────────────────────────────────────────────────────────────
n_avant = len(lib)
lib = lib.drop_duplicates(subset=["ORDER_SHOP_CD"])
log(f"Doublons ORDER_SHOP_CD supprimes dans LIB_BQT : {n_avant - len(lib):,}")

# ── 5.2 Nettoyage clé de jointure ────────────────────────────────────────────
lib["ORDER_SHOP_CD"] = lib["ORDER_SHOP_CD"].astype(str).str.strip()

# ── 5.3 Nettoyage MKT_CHANNL_LB ──────────────────────────────────────────────
lib["MKT_CHANNL_LB"] = lib["MKT_CHANNL_LB"].astype(str).str.strip()
lib["MKT_CHANNL_LB"] = lib["MKT_CHANNL_LB"].replace({"nan": np.nan, "": np.nan})
log(f"MKT_CHANNL_LB : {lib['MKT_CHANNL_LB'].value_counts(dropna=False).to_dict()}")

# ── 5.4 Filtre : garder uniquement les boutiques physiques ───────────────────
# LIB_BQT contient TOUTES les boutiques (physiques + online + porte-à-porte...)
# On ne garde que celles qui sont aussi dans COORD_BQT (= boutiques physiques)
codes_physiques = set(coord["ORDER_SHOP_CD"].unique())
n_avant = len(lib)
lib_physique = lib[lib["ORDER_SHOP_CD"].isin(codes_physiques)].copy()
log(f"LIB_BQT filtre boutiques physiques uniquement : {n_avant} -> {len(lib_physique)} | "
    f"(boutiques non-physiques exclues : {n_avant - len(lib_physique)})")


# =============================================================================
# ETAPE 6 — JOINTURE DES TROIS TABLES
# =============================================================================
separateur("ETAPE 6 — JOINTURE")

n_avant = len(df)

# Jointure 1 : fichier principal + coordonnées boutiques
df_joint = df.merge(coord, on="ORDER_SHOP_CD", how="left")
log(f"Apres jointure avec COORD_BQT : {len(df_joint):,} lignes")
n_sans_coord = df_joint["Coord_L2E_II_X"].isnull().sum()
log(f"  Lignes sans coordonnees boutique : {n_sans_coord:,} "
    f"({n_sans_coord/len(df_joint)*100:.1f}%) — boutiques inconnues ou fermees")

# Jointure 2 : + libellés boutiques
df_joint = df_joint.merge(lib_physique[["ORDER_SHOP_CD", "MKT_CHANNL_LB"]],
                           on="ORDER_SHOP_CD", how="left")
log(f"Apres jointure avec LIB_BQT : {len(df_joint):,} lignes")
n_sans_lib = df_joint["MKT_CHANNL_LB"].isnull().sum()
log(f"  Lignes sans libelle boutique : {n_sans_lib:,} ({n_sans_lib/len(df_joint)*100:.1f}%)")

# Vérification : pas de création de lignes fantômes
if len(df_joint) != n_avant:
    log(f"  ATTENTION : jointure a change le nb de lignes ({n_avant} -> {len(df_joint)}) !")
    log(f"  Cela peut indiquer des ORDER_SHOP_CD dupliques dans COORD_BQT ou LIB_BQT")
else:
    log(f"  OK : nb de lignes stable apres jointure ({len(df_joint):,})")


# =============================================================================
# ETAPE 7 — CREATION DES NOUVELLES VARIABLES
# =============================================================================
separateur("ETAPE 7 — CREATION DES NOUVELLES VARIABLES")

# ── 7.1 age_client ───────────────────────────────────────────────────────────
annee_ref = df_joint["PERIOD_YYYY"].max()   # année de référence = année max des données
df_joint["age_client"] = annee_ref - df_joint["PERSON_BIRTH_DT_Year"]
log(f"age_client cree (annee ref = {annee_ref}) | min={df_joint['age_client'].min():.0f}, max={df_joint['age_client'].max():.0f}")

# ── 7.2 tranche_age ──────────────────────────────────────────────────────────
bins   = [0, 25, 35, 50, 65, 150]
labels = ["18-25", "26-35", "36-50", "51-65", "65+"]
df_joint["tranche_age"] = pd.cut(df_joint["age_client"], bins=bins, labels=labels, right=True)
log(f"tranche_age cree : {df_joint['tranche_age'].value_counts().to_dict()}")

# ── 7.3 trimestre ────────────────────────────────────────────────────────────
df_joint["trimestre"] = df_joint["PERIOD_MM"].apply(
    lambda m: f"T{int((m-1)//3 + 1)}" if pd.notna(m) else np.nan
)
log(f"trimestre cree : {df_joint['trimestre'].value_counts().to_dict()}")

# ── 7.4 Conversion coordonnées Lambert II -> latitude/longitude ──────────────
# Les coordonnées dans ce dataset sont en Lambert II étendu (projection française)
# On les convertit en WGS84 (lat/lon) pour pouvoir calculer des distances réelles
print("\n[7.4] Conversion Lambert II -> WGS84 (lat/lon)...")
try:
    from pyproj import Transformer
    transformer = Transformer.from_crs("EPSG:27572", "EPSG:4326", always_xy=True)

    # Conversion pour les boutiques
    coord_x_bqt = df_joint["Coord_L2E_II_X"].values
    coord_y_bqt = df_joint["Coord_L2E_II_Y"].values
    lon_bqt, lat_bqt = transformer.transform(coord_x_bqt, coord_y_bqt)
    df_joint["lat_boutique"] = np.where(np.isfinite(lat_bqt), lat_bqt, np.nan)
    df_joint["lon_boutique"] = np.where(np.isfinite(lon_bqt), lon_bqt, np.nan)

    # Conversion pour les clients
    coord_x_cli = df_joint["Coord_L2E_II_X"].values
    coord_y_cli = df_joint["Coord_L2E_II_Y"].values
    lon_cli, lat_cli = transformer.transform(coord_x_cli, coord_y_cli)
    df_joint["lat_client"] = np.where(np.isfinite(lat_cli), lat_cli, np.nan)
    df_joint["lon_client"] = np.where(np.isfinite(lon_cli), lon_cli, np.nan)

    log("Coordonnees converties en WGS84 (lat/lon)")

except ImportError:
    log("pyproj non installe — coordonnees Lambert II conservees telles quelles")
    log("=> Pour installer : pip install pyproj")
    df_joint["lat_boutique"] = np.nan
    df_joint["lon_boutique"] = np.nan
    df_joint["lat_client"]   = np.nan
    df_joint["lon_client"]   = np.nan

# ── 7.5 distance_km (formule Haversine) ──────────────────────────────────────
print("\n[7.5] Calcul de la distance client/boutique (Haversine)...")

def haversine_vectorise(lat1, lon1, lat2, lon2):
    """Calcule la distance en km entre deux points GPS (vectorisé)."""
    R = 6371  # rayon de la Terre en km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return R * 2 * np.arcsin(np.sqrt(a))

if "lat_boutique" in df_joint.columns and df_joint["lat_boutique"].notna().any():
    df_joint["distance_km"] = haversine_vectorise(
        df_joint["lat_client"].values,
        df_joint["lon_client"].values,
        df_joint["lat_boutique"].values,
        df_joint["lon_boutique"].values
    )
    # Supprimer les distances aberrantes (> 500km ou négatives)
    df_joint.loc[df_joint["distance_km"] > 500, "distance_km"] = np.nan
    df_joint.loc[df_joint["distance_km"] < 0,   "distance_km"] = np.nan
    moy = df_joint["distance_km"].mean()
    med = df_joint["distance_km"].median()
    log(f"distance_km cree | moyenne={moy:.1f} km | mediane={med:.1f} km")
else:
    df_joint["distance_km"] = np.nan
    log("distance_km non calculee (coordonnees manquantes ou pyproj absent)")

# ── 7.6 rayon_chalandise ─────────────────────────────────────────────────────
bins_dist   = [0, 5, 20, 50, 9999]
labels_dist = ["<5km", "5-20km", "20-50km", ">50km"]
df_joint["rayon_chalandise"] = pd.cut(
    df_joint["distance_km"], bins=bins_dist, labels=labels_dist, right=True
)
log(f"rayon_chalandise cree : {df_joint['rayon_chalandise'].value_counts().to_dict()}")

# ── 7.7 Libellés lisibles pour NET_ZONE_IN et DENSE_ZONE_IN ─────────────────
NET_ZONE_MAP = {
    "123": "SFR+FREE+BOUYGUES",
    "12":  "SFR+FREE",
    "13":  "SFR+BOUYGUES",
    "1":   "SFR seul",
    "23":  "FREE+BOUYGUES",
    "2":   "FREE seul",
    "3":   "BOUYGUES seul",
}
DENSE_MAP = {
    "0": "Non Dense",
    "1": "Dense",
    "2": "ZTHD Non cable",
    "3": "ZTHD Cable",
}

if "NET_ZONE_IN" in df_joint.columns:
    df_joint["concurrence_fibre"] = df_joint["NET_ZONE_IN"].astype(str).map(NET_ZONE_MAP)
    log(f"concurrence_fibre cree : {df_joint['concurrence_fibre'].value_counts(dropna=False).to_dict()}")

if "DENSE_ZONE_IN" in df_joint.columns:
    df_joint["densite_zone"] = df_joint["DENSE_ZONE_IN"].astype(str).map(DENSE_MAP)
    log(f"densite_zone cree : {df_joint['densite_zone'].value_counts(dropna=False).to_dict()}")


# =============================================================================
# ETAPE 8 — CONTROLES QUALITE FINAUX
# =============================================================================
separateur("ETAPE 8 — CONTROLES QUALITE FINAUX")

print("\n[8.1] Resume du dataset final :")
print(f"  Nb lignes          : {len(df_joint):,}")
print(f"  Nb colonnes        : {len(df_joint.columns)}")
print(f"  Boutiques uniques  : {df_joint['ORDER_SHOP_CD'].nunique()}")
print(f"  Annees couvertes   : {sorted(df_joint['PERIOD_YYYY'].dropna().unique().tolist())}")
print(f"  Zone               : {df_joint['ZONE'].unique().tolist()}")

print("\n[8.2] Valeurs manquantes dans le dataset final :")
nulls_final = df_joint.isnull().sum()
nulls_final_pct = (nulls_final / len(df_joint) * 100).round(2)
diag_final = pd.DataFrame({"Nulls": nulls_final, "Pct (%)": nulls_final_pct})
diag_final = diag_final[diag_final["Nulls"] > 0].sort_values("Pct (%)", ascending=False)
print(diag_final.to_string())

print("\n[8.3] Liste des colonnes finales :")
for col in df_joint.columns:
    print(f"  - {col} ({df_joint[col].dtype})")


# =============================================================================
# ETAPE 9 — SAUVEGARDE
# =============================================================================
separateur("ETAPE 9 — SAUVEGARDE")

fichier_sortie = "base_travail_SUD_EST_propre.csv"
df_joint.to_csv(fichier_sortie, index=False, encoding="utf-8-sig")
log(f"Fichier sauvegarde : {fichier_sortie}")
log(f"Taille finale : {len(df_joint):,} lignes x {len(df_joint.columns)} colonnes")

print(f"\n  Apercu des 3 premieres lignes :")
print(df_joint.head(3).to_string())


# =============================================================================
# JOURNAL DE NETTOYAGE (a copier dans tes annexes)
# =============================================================================
separateur("JOURNAL DE NETTOYAGE COMPLET")
journal_path = "journal_nettoyage.txt"
with open(journal_path, "w", encoding="utf-8") as f:
    f.write("JOURNAL DE NETTOYAGE — TELECOM GROUPE 5 SUD-EST\n")
    f.write("=" * 60 + "\n\n")
    for i, line in enumerate(journal, 1):
        f.write(f"{i:03d}. {line}\n")

print(f"\n  Journal sauvegarde dans : {journal_path}")
print(f"  Total decisions documentees : {len(journal)}")
print("\n" + "="*65)
print("  NETTOYAGE TERMINE AVEC SUCCES !")
print("  Fichier de sortie : base_travail_SUD_EST_propre.csv")
print("="*65)
```

```python
# ============================================================
#   KPI COMPARAISONS – TELECOM SUD-EST | Groupe 5
#   Fichier : Fichier_Principal_pedago_anonym.csv
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

# ─── CONFIGURATION ───────────────────────────────────────────
FICHIER   = "Fichier_Principal_pedago_anonym.csv"
SEP       = ";"
ENCODING  = "utf-8"
ZONE      = "5_SUD_EST"
PALETTE   = ["#2ecc71", "#e74c3c", "#3498db", "#f39c12", "#9b59b6",
             "#1abc9c", "#e67e22", "#34495e", "#e91e63", "#00bcd4"]
# ─────────────────────────────────────────────────────────────

# Lecture du fichier : on tente d'abord avec le séparateur configuré (SEP).
# Si le DataFrame n'a qu'une seule colonne (cas fréquent si le séparateur est incorrect),
# on retente automatiquement avec une détection du séparateur.
try:
    df = pd.read_csv(FICHIER, sep=SEP, encoding=ENCODING, low_memory=False)
    # cas où tout le contenu est dans une seule colonne (mauvais séparateur)
    if df.shape[1] == 1:
        # Essai avec la virgule si ce n'était pas déjà le séparateur
        if SEP != ',':
            try:
                df_tmp = pd.read_csv(FICHIER, sep=',', encoding=ENCODING, low_memory=False)
                if df_tmp.shape[1] > 1:
                    df = df_tmp
            except Exception:
                pass
        # Dernier recours : laisser pandas détecter automatiquement (engine='python')
        if df.shape[1] == 1:
            df = pd.read_csv(FICHIER, sep=None, engine='python', encoding=ENCODING)
except Exception:
    # Si la lecture initiale échoue, on retente avec détection automatique
    df = pd.read_csv(FICHIER, sep=None, engine='python', encoding=ENCODING)

# Nettoyage des noms de colonnes (espaces superflus)
df.columns = df.columns.str.strip()

# Normalisation du nom de la colonne ZONE (gestion de casse/espaces)
if "ZONE" not in df.columns:
    for c in df.columns:
        if c.strip().upper() == "ZONE":
            df.rename(columns={c: "ZONE"}, inplace=True)
            break

# Si la colonne ZONE existe, on filtre ; sinon on continue sans filtrer et on émet un message informatif.
if "ZONE" in df.columns:
    df = df[df["ZONE"] == ZONE].copy()
else:
    print(f"⚠️ La colonne 'ZONE' est introuvable dans le fichier {FICHIER}. Aucun filtrage par zone appliqué.")

# Colonnes temporelles
if "PERIOD_YYYY" in df.columns:
    df["annee"] = df["PERIOD_YYYY"].astype(str)
if "PERIOD_MM" in df.columns:
    df["mois"] = df["PERIOD_MM"].astype(int)

# Tranche d'âge
if "PERSON_BIRTH_DT_year" in df.columns:
    df["age"] = 2024 - pd.to_numeric(df["PERSON_BIRTH_DT_year"], errors="coerce")
    bins  = [0, 25, 35, 45, 55, 65, 100]
    labs  = ["<25", "25-34", "35-44", "45-54", "55-64", "65+"]
    df["tranche_age"] = pd.cut(df["age"], bins=bins, labels=labs, right=False)


# ═══════════════════════════════════════════════════════════════
#  GRAPHIQUE 1 – Volume d'acquisitions par LINE_TYPE × Année
#  → Quels produits progressent / régressent dans le temps ?
# ═══════════════════════════════════════════════════════════════
if "LINE_TYPE" in df.columns and "annee" in df.columns:
    pivot = (df.groupby(["annee", "LINE_TYPE"])
               .size()
               .reset_index(name="volume"))
    annees    = sorted(pivot["annee"].unique())
    line_types = pivot["LINE_TYPE"].unique()
    x     = np.arange(len(annees))
    width = 0.8 / len(line_types)

    fig, ax = plt.subplots(figsize=(12, 6))
    for i, lt in enumerate(line_types):
        data = pivot[pivot["LINE_TYPE"] == lt].set_index("annee").reindex(annees, fill_value=0)
        bars = ax.bar(
            x + i * width,
            data["volume"],
            width=width * 0.9,
            label=lt,
            color=PALETTE[i % len(PALETTE)],
            edgecolor="white", linewidth=0.5
        )
        for b in bars:
            h = b.get_height()
            if h > 0:
                ax.text(b.get_x() + b.get_width() / 2, h + h * 0.02,
                        f"{int(h):,}", ha="center", va="bottom", fontsize=7, color="#333")

    ax.set_xticks(x + width * (len(line_types) - 1) / 2)
    ax.set_xticklabels(annees, fontsize=10)
    ax.set_title("Volume d'acquisitions par produit (LINE_TYPE) et par année",
                 fontsize=13, fontweight="bold", pad=15)
    ax.set_ylabel("Nombre d'acquisitions")
    ax.legend(title="LINE_TYPE", bbox_to_anchor=(1.01, 1), loc="upper left", frameon=False)
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{int(x):,}"))
    ax.spines[["top", "right"]].set_visible(False)
    plt.tight_layout()
    plt.savefig("g1_linetype_annee.png", dpi=150)
    plt.show()
    print("✅ g1_linetype_annee.png")


# ═══════════════════════════════════════════════════════════════
#  GRAPHIQUE 2 – Saisonnalité mensuelle par LINE_TYPE
#  → Y a-t-il des mois de pointe selon le type de produit ?
# ═══════════════════════════════════════════════════════════════
if "LINE_TYPE" in df.columns and "mois" in df.columns:
    mois_labels = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun",
                   "Jul", "Aoû", "Sep", "Oct", "Nov", "Déc"]

    pivot_m = (df.groupby(["mois", "LINE_TYPE"])
                 .size()
                 .reset_index(name="volume"))

    fig, ax = plt.subplots(figsize=(12, 6))
    for i, lt in enumerate(pivot_m["LINE_TYPE"].unique()):
        data = (pivot_m[pivot_m["LINE_TYPE"] == lt]
                .set_index("mois")
                .reindex(range(1, 13), fill_value=0))
        ax.plot(
            range(1, 13),
            data["volume"],
            marker="o", markersize=5,
            linewidth=2,
            label=lt,
            color=PALETTE[i % len(PALETTE)]
        )

    ax.set_xticks(range(1, 13))
    ax.set_xticklabels(mois_labels)
    ax.set_title("Saisonnalité mensuelle des acquisitions par LINE_TYPE",
                 fontsize=13, fontweight="bold", pad=15)
    ax.set_ylabel("Nombre d'acquisitions")
    ax.legend(title="LINE_TYPE", bbox_to_anchor=(1.01, 1), loc="upper left", frameon=False)
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{int(x):,}"))
    ax.spines[["top", "right"]].set_visible(False)
    plt.tight_layout()
    plt.savefig("g2_saisonnalite.png", dpi=150)
    plt.show()
    print("✅ g2_saisonnalite.png")


# ═══════════════════════════════════════════════════════════════
#  GRAPHIQUE 3 – Profil d'âge par LINE_TYPE (barres 100% empilées)
#  → Quel segment d'âge achète quel produit ?
# ═══════════════════════════════════════════════════════════════
if "LINE_TYPE" in df.columns and "tranche_age" in df.columns:
    pivot_a = (df.groupby(["LINE_TYPE", "tranche_age"])
                 .size()
                 .reset_index(name="volume"))

    totaux = pivot_a.groupby("LINE_TYPE")["volume"].transform("sum")
    pivot_a["pct"] = pivot_a["volume"] / totaux * 100

    line_types_a = pivot_a["LINE_TYPE"].unique()
    tranches     = labs

    fig, ax = plt.subplots(figsize=(12, 6))
    bottoms = np.zeros(len(line_types_a))

    age_colors = ["#d35400", "#e67e22", "#f1c40f", "#2ecc71", "#3498db", "#9b59b6"]
    for j, tr in enumerate(tranches):
        vals = []
        for lt in line_types_a:
            row = pivot_a[(pivot_a["LINE_TYPE"] == lt) & (pivot_a["tranche_age"] == tr)]
            vals.append(row["pct"].values[0] if not row.empty else 0)
        bars = ax.bar(
            line_types_a, vals, bottom=bottoms,
            color=age_colors[j], label=tr, edgecolor="white", linewidth=0.5
        )
        for b, v in zip(bars, vals):
            if v > 5:
                ax.text(b.get_x() + b.get_width() / 2,
                        b.get_y() + b.get_height() / 2,
                        f"{v:.0f}%", ha="center", va="center",
                        fontsize=7, color="white", fontweight="bold")
        bottoms += np.array(vals)

    ax.set_title("Répartition des tranches d'âge par LINE_TYPE (barres 100% empilées)",
                 fontsize=13, fontweight="bold", pad=15)
    ax.set_ylabel("% d'acquisitions")
    ax.set_ylim(0, 105)
    ax.legend(title="Tranche d'âge", bbox_to_anchor=(1.01, 1), loc="upper left", frameon=False)
    ax.spines[["top", "right"]].set_visible(False)
    plt.tight_layout()
    plt.savefig("g3_age_linetype.png", dpi=150)
    plt.show()
    print("✅ g3_age_linetype.png")


# ═══════════════════════════════════════════════════════════════
#  GRAPHIQUE 4 – Top 15 boutiques : volume × croissance Y/Y
#  → Scatter "performance vs dynamisme" — quelles boutiques sont
#    en déclin malgré un fort volume ?
# ═══════════════════════════════════════════════════════════════
if "ORDER_SHOP_CD" in df.columns and "annee" in df.columns:
    annees_dispo = sorted(df["annee"].dropna().unique())
    if len(annees_dispo) >= 2:
        an_n   = annees_dispo[-1]
        an_n1  = annees_dispo[-2]

        vol_n  = df[df["annee"] == an_n].groupby("ORDER_SHOP_CD").size().rename("vol_n")
        vol_n1 = df[df["annee"] == an_n1].groupby("ORDER_SHOP_CD").size().rename("vol_n1")

        perf = pd.concat([vol_n, vol_n1], axis=1).fillna(0)
        perf["croissance"] = (perf["vol_n"] - perf["vol_n1"]) / perf["vol_n1"].replace(0, np.nan) * 100
        perf = perf.dropna(subset=["croissance"])
        top15 = perf.nlargest(15, "vol_n")

        fig, ax = plt.subplots(figsize=(11, 7))
        couleurs = ["#e74c3c" if c < 0 else "#2ecc71" for c in top15["croissance"]]
        sc = ax.scatter(
            top15["vol_n"],
            top15["croissance"],
            c=couleurs,
            s=top15["vol_n"] / top15["vol_n"].max() * 400 + 40,
            alpha=0.8, edgecolors="white", linewidth=0.8
        )
        for shop, row in top15.iterrows():
            ax.annotate(
                str(shop),
                (row["vol_n"], row["croissance"]),
                textcoords="offset points", xytext=(6, 4),
                fontsize=7, color="#333"
            )

        ax.axhline(0, color="#aaa", linewidth=1, linestyle="--")
        ax.axvline(top15["vol_n"].median(), color="#ccc", linewidth=1,
                   linestyle=":", label=f"Médiane volume")
        ax.set_title(f"Top 15 boutiques : volume ({an_n}) × croissance vs {an_n1}\n"
                     "🔴 Déclin  🟢 Croissance  — taille du point ∝ volume",
                     fontsize=12, fontweight="bold", pad=12)
        ax.set_xlabel(f"Volume d'acquisitions {an_n}")
        ax.set_ylabel(f"Croissance vs {an_n1} (%)")
        ax.legend(frameon=False, fontsize=8)
        ax.spines[["top", "right"]].set_visible(False)
        plt.tight_layout()
        plt.savefig("g4_scatter_boutiques.png", dpi=150)
        plt.show()
        print("✅ g4_scatter_boutiques.png")


# ═══════════════════════════════════════════════════════════════
#  GRAPHIQUE 5 – Segment Geolife × LINE_TYPE (heatmap)
#  → Quel segment socio-démo achète quoi ?
# ═══════════════════════════════════════════════════════════════
if "Geolife" in df.columns and "LINE_TYPE" in df.columns:
    pivot_g = (df.groupby(["Geolife", "LINE_TYPE"])
                 .size()
                 .unstack(fill_value=0))
    # Normalisation par ligne (profil d'achat du segment)
    pivot_pct = pivot_g.div(pivot_g.sum(axis=1), axis=0) * 100

    fig, ax = plt.subplots(figsize=(12, max(6, len(pivot_pct) * 0.45)))
    im = ax.imshow(pivot_pct.values, aspect="auto", cmap="YlOrRd")

    ax.set_xticks(range(len(pivot_pct.columns)))
    ax.set_xticklabels(pivot_pct.columns, rotation=35, ha="right", fontsize=9)
    ax.set_yticks(range(len(pivot_pct.index)))
    ax.set_yticklabels(pivot_pct.index, fontsize=8)

    for i in range(pivot_pct.shape[0]):
        for j in range(pivot_pct.shape[1]):
            val = pivot_pct.values[i, j]
            color = "white" if val > 50 else "#333"
            ax.text(j, i, f"{val:.0f}%", ha="center", va="center", fontsize=7, color=color)

    plt.colorbar(im, ax=ax, label="% du mix produit du segment")
    ax.set_title("Mix produit (LINE_TYPE) par segment Geolife (%)",
                 fontsize=13, fontweight="bold", pad=15)
    plt.tight_layout()
    plt.savefig("g5_heatmap_geolife.png", dpi=150)
    plt.show()
    print("✅ g5_heatmap_geolife.png")

print("\n🎯 Tous les graphiques ont été générés.")
```
