## Exercise C1 – Flight Prices (filter + sort)

Filter the flights where:
- `origin == 'MRS'`
- `price_eur <= 120`

Then sort by `price_eur` ascending.

Store the resulting DataFrame in `mrs_cheapest`.

```python
import pandas as pd

flights = pd.DataFrame({
    "origin": ["MRS","MRS","LYS","MRS","CDG","MRS"],
    "dest":   ["CDG","NCE","CDG","BCN","MRS","NCE"],
    "price_eur":[110, 140, 95, 120, 105, 80],
    "airline":["A","B","A","C","B","A"]
})

mrs_cheapest = flights.loc[(flights["origin"]=="MRS") & (flights["price_eur"]<=120)].sort_values("price_eur")
mrs_cheapest
```

Use `value_counts()` on `platform`.

Store:
- the counts Series in `platform_counts`
- the most frequent platform name in `top_platform` (no `idxmax`)


```python
import pandas as pd

installs = pd.DataFrame({
    "user": ["u1","u2","u3","u4","u5","u6","u7"],
    "platform": ["iOS","Android","Android","iOS","Android","Android","iOS"]
})

platform_counts= installs["platform"].value_counts()
top_platform = platform_counts.idxmax()
top_platform2 = platform_counts.sort_values(ascending=False).index[0]
top_platform2
```

## Exercise C5 – Daily Steps (missing values)

1) Count how many missing values are in `steps` and store it in `n_missing_steps`.
2) Create `steps_filled` where missing `steps` are filled with the median.


```python
import pandas as pd
import numpy as np

steps = pd.DataFrame({
    "day": ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"],
    "steps": [8000, None, 12000, 5000, None, 15000, 7000]
})
n_missing_steps = steps["steps"].isna().sum()
steps_filled = steps.copy()
steps_filled["steps"] = steps_filled["steps"].fillna(steps_filled["steps"].mean())
steps_filled
```

```python
import pandas as pd
import numpy as np

steps = pd.DataFrame({
    "day": ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"],
    "steps": [8000, None, 12000, 5000, None, 15000, 7000]
})
n_missing_steps = int(steps["steps"].isna().sum())

steps_filled = steps.copy()


steps_filled["steps"] = steps_filled["steps"].fillna(steps_filled["steps"].median())
n_missing_steps, steps_filled
```

## Exercise C6 – Ticket Priority (map)

Create a new column `priority_num` from `priority` using this mapping:
`{'low':1, 'medium':2, 'high':3}`

Store the resulting DataFrame in `tickets_scored`.

```python
import pandas as pd

tickets = pd.DataFrame({
    "id": [1,2,3,4,5],
    "priority": ["low","high","medium","low","high"]
})

mapping = {"low": 1, "medium": 2, "high": 3}
tickets["priority_num"]= tickets["priority"].map(mapping)
tickets
```

## Exercise C9 – Index alignment (Series + DataFrame)

1) Create a DataFrame `prices` from `data_dict`.
2) Set `country` as the index.
3) Create a Series `vat` with VAT rates for:
- France: 0.20
- Germany: 0.19
- Spain: 0.21
4) Create a new column `price_with_vat` using **index alignment**:
`price_with_vat = price_eur * (1 + vat)`

Store the final DataFrame in `prices_vat`.

```python
import pandas as pd

data_dict = {
    "country": ["France", "Germany", "Spain", "Italy"],
    "price_eur": [10.0, 12.0, 8.0, 9.0]
}
prices = pd.DataFrame(data_dict).set_index("country")
vat = pd.Series({"France" : 0.20, 'Germany':0.19,"Spain":0.21})
prices["price_with_vat"]=prices["price_eur"]*(1+vat)
prices
```
